```python
import math
from math import *
import socket 
import threading 
import PIL
from PIL import ImageTk,Image
from tkinter import *
from tkinter import font
import requests
from tkinter import ttk 
from tkinter import filedialog
FORMAT = "utf-8"

host=socket.gethostname()
client = socket.socket(socket.AF_INET,socket.SOCK_STREAM) 
client.connect((host,5000)) 

class ap: 
    private_t=0
    public_t=0
    def __init__(self): 

        self.Window = Tk() 
        self.Window.withdraw() 
        

        self.ap_gui = Toplevel()  
        self.ap_gui.title("chat") 
        self.ap_gui.resizable(width = False,height = False) 
        self.ap_gui.configure(width = 1000,height = 600) 
        
        self.lablewelcome = Label(self.ap_gui, text = "Please Register to continue", 
                    justify = CENTER, font = "Helvetica 14 bold") 
        self.lablewelcome.place(relheight = 0.15,  relx = 0.7, rely = 0.009) 
        
   
        self.labelName = Label(self.ap_gui, text = "Name: ", font = "Helvetica 12 bold") 
        self.labelName.place(relheight = 0.2,  relx = 0.6,  rely = 0.6)
        
 
                 
        self.entryName = Entry(self.ap_gui, font = "Helvetica 14") 
        self.entryName.place(relwidth = 0.3,relheight = 0.08,relx = 0.68,rely = 0.66) 
        self.entryName.focus()         

        
        self.labelURL = Label(self.ap_gui,  text = "URL: ", font = "Helvetica 12 bold") 
        self.labelURL.place(relheight = 0.12, relx = 0.6, rely = 0.54)

        self.entryURL = Entry(self.ap_gui, font = "Helvetica 14") 
        self.entryURL.place(relwidth = 0.3,relheight = 0.08,relx = 0.68, rely = 0.56) 
        self.entryURL.focus() 
        
        
        self.labelPR1 = Label(self.ap_gui,text = "prime 1: ", font = "Helvetica 12 bold") 
        self.labelPR1.place(relheight = 0.2,  relx = 0.6,  rely = 0.73)
        
        self.entryPR1 = Entry(self.ap_gui,font = "Helvetica 14") 
        self.entryPR1.place(relwidth = 0.1,relheight = 0.08, relx = 0.68, rely = 0.79) 
        self.entryPR1.focus() 
        
    
        self.labelPR2= Label(self.ap_gui, text = "prime 2: ", font = "Helvetica 12 bold")
        self.labelPR2.place(relheight = 0.2,  relx = 0.8, rely = 0.74)

        self.entryPR2 = Entry(self.ap_gui,font = "Helvetica 14") 
        self.entryPR2.place(relwidth = 0.1,relheight = 0.08,  relx = 0.88,  rely = 0.79) 
        self.entryPR2.focus() 
        
        self.BTRegister = Button(self.ap_gui,text = "Register", font = "Helvetica 14 bold", 
                        command = lambda: self.midenmethod(self.entryName.get(),self.entryPR1.get(),self.entryPR2.get())) 
        self.BTRegister.place(relx =0.68, rely =0.9)
        self.Change_photo = Button(self.ap_gui,text = "Change photo", font = "Helvetica 14 bold",
                                   command = lambda: self.dowleod(self.entryURL.get())) 
        self.Change_photo.place(relx = 0.83, rely =0.9)
        self.Window.mainloop()
        
    def dowleod(self,url):
        if url!="":
            self.entryURL.delete(0, END)
            response = requests.get(url)
            file = open("sample2_image.jpg", "wb")
            file.write(response.content)
            file.close()
            self.lblmage = Label(self.ap_gui) 
            self.lblmage.place(relx = 0.7, rely = 0.12)
            self.img=PIL.Image.open("sample2_image.jpg")
            self.img.thumbnail((250,250))
            self.img=ImageTk.PhotoImage( self.img)
            self.lblmage.configure(image= self.img)
            self.lblmage.image= self.img 
   
        
    def midenmethod(self, name ,p1,p2):
        if name!="":
            self.entryPR1.delete(0, END)
            self.entryPR2.delete(0, END)
            self.entryName.delete(0, END)
            self.roomgui(name,p1,p2) 
            self.entryName.config(state = DISABLED)
            self.entryPR1.config(state = DISABLED)
            self.entryPR2.config(state = DISABLED)
            rcv = threading.Thread(target=self.receive_msg) 
            rcv.start() 
        
    def roomgui(self,name,p1,p2):
        self.name = name 
        self.p =int(p1)
        self.q = int(p2)
        self.n = self.p *self.q
        self.r= (self.p-1)*(self.q-1)

        def egcd(e,r):
            while(r!=0):
                e,r=r,e%r
            return e

        def eugcd(e,r):
            for i in range(1,r):
                while(e!=0):
                        a,b=r//e,r%e
                        r=e
                        e=b

        def eea(a,b):
            if(a%b==0):
                return(b,0,1)
            else:
                gcd,s,t = eea(b,a%b)
                s = s-((a//b) * t)
                return(gcd,t,s)
    
        def mult_inv(e,r):
            gcd,s,_=eea(e,r)
            if(gcd!=1):
                return None
            else:
                return s%r

        for i in range(1,1000):
            if(egcd(i,self.r)==1):
                self.e=i

        eugcd(self.e,self.r)

        self.d = mult_inv(self.e,self.r)
        self.public = (self.e,self.n)
        self.private = (self.d,self.n)
        self.public_t= self.public
        self.private_t=self.private 
  


        self.labelHead = Label(self.ap_gui, bg = "#17202A", fg = "#EAECEE", text =  self.name, 
                            font = "Helvetica 13 bold", pady = 5) 
        self.labelHead.place(relwidth = 1) 
        

        self.line = Label(self.ap_gui, width = 100,bg = "#ABB2B9")
        self.line.place(relwidth = 1, rely = 0.05,relheight = 0.012) 
       
        self.TextRoomMesage = Text(self.ap_gui,  width = 4, height = 2,   bg = "#010003", fg = "#EAECEE",
                            font = "Helvetica 14", padx = 1, pady = 1) 
        self.TextRoomMesage.place(relheight = 0.74,relwidth = 0.52,rely = 0.08,relx = 0.04 ) 
        

        self.labelBottom = Label(self.ap_gui,bg = "#ABB2B9",height = 80) 
        self.labelBottom.place(relwidth = 0.52, rely = 0.825,relx = 0.04)
        
        self.entryMsg = Entry(self.labelBottom, bg = "#010003",fg = "#EAECEE", font = "Helvetica 13") 
        self.entryMsg.place(relwidth = 0.74,relheight = 0.06, rely = 0.008, relx = 0.011) 
        self.entryMsg.focus() 
        
        self.buttonMsg = Button(self.labelBottom,  text = "Send", font = "Helvetica 14 bold", 
                                width = 20,  fg="#FFFFFF",bg = "#AAAAAA", 
                                command = lambda : self.sdBT(self.entryMsg.get(), self.public)) 
        self.buttonMsg.place(relx = 0.77,rely = 0.008,relheight = 0.06,relwidth = 0.22)
        
        self.TextRoomMesage.config(cursor = "arrow") 

        scrollbar = Scrollbar(self.TextRoomMesage) 
        scrollbar.place(relheight = 1, relx = 0.974) 
        scrollbar.config(command = self.TextRoomMesage.yview) 
        
        self.TextRoomMesage.config(state = DISABLED) 
        
        
    def encrypt(self,pub_key,n_text):
        e,n=pub_key
        x=[] ;m=0;
        for i in n_text:
            if(i.isupper()):
                m = ord(i)-65
                c=(m**e)%n
                x.append(c)
            elif(i.islower()):               
                m= ord(i)-97
                c=(m**e)%n
                x.append(c)
            elif(i.isspace()):
                spc=400
                x.append(400)
            elif(i==':'):
                spc=1000
                x.append(1000)
        return x


    def decrypt(self,priv_key,c_text):
        d,n=priv_key
        txt=c_text
        x=''
        m=0
        for i in txt:
            if(i==400):
                x+=' '
            elif(i==1000):
                x+=':'
            else:
                m=(int(i)**d)%n
                m+=65
                c=chr(m)
                x+=c
        return x


    def sdBT(self, msg,public): 
        self.TextRoomMesage.config(state = DISABLED) 
        self.msg=msg
        self.entryMsg.delete(0, END) 
        snd= threading.Thread(target = self.sendMessage) 
        snd.start() 

 
    def receive_msg(self): 
        self.f=0
        while True: 
            self.f+=1
            try: 
                message = client.recv(1024).decode(FORMAT)
                if message == 'NAME': 
                    client.send(self.name.encode(FORMAT)) 
                else:
                    if message[0]=='[':
                        z=[]
                        s=""
                        for i in message:
                            if i!=','and i!='['and i!=']'and i!=' ':
                                s+=i
                            else:
                                if s!="":
                                    z.append(int(s))
                                    s=""
        
                        message_decrypt=self.decrypt(self.private_t,z)
                        self.TextRoomMesage.config(state = NORMAL) 
                        self.TextRoomMesage.insert(END,message_decrypt+"\n\n")     
                        self.TextRoomMesage.config(state = DISABLED) 
                        self.TextRoomMesage.see(END)
                    else:  
                        self.TextRoomMesage.config(state = NORMAL) 
                        self.TextRoomMesage.insert(END,message+"\n\n")     
                        self.TextRoomMesage.config(state = DISABLED) 
                        self.TextRoomMesage.see(END) 
            except:  
                print("not joined the chat!!") 
                client.close() 
                break
                
 
    def sendMessage(self): 
        self.TextRoomMesage.config(state=DISABLED) 
        while True: 
            message =self.name+" : "+self.msg
            message_encrypt=self.encrypt(self.public_t,message)
            message_encrypt = (f"{message_encrypt}") 
            client.send(message_encrypt.encode(FORMAT))
            break            

g = ap() 
```
